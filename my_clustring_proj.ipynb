```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from my_clustring import KMeans
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
```

```python
data = pd.read_csv("data.csv")
data
```

```python
x = data.drop("label", axis=1)
y = data["label"]
```

```python
plt.scatter(x["feature1"], x["feature2"])
plt.show()
```

```python
k = KMeans(n_clusters=3)
k.fit(x.values)
```

```python
k.labels
```

```python
predicted_cluster = k.predict(x.values)
plt.scatter(x["feature1"], x["feature2"], c=predicted_cluster)
```

```python
for i, idx in enumerate(k.labels):
    point = x.values[idx].T
    plt.scatter(*point)

for i in k.centroids:
    plt.scatter(*i, marker="x", c="black")
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, init="k-means++", tol=1e4)
model.fit(x)
```

```python
predicted_cluster = model.predict(x)
plt.scatter(x["feature1"], x["feature2"], c=predicted_cluster)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c="red", marker="x")
```

```python
model.cluster_centers_
```

```python
max_clusres = 10
distances = []
for i in range(1, max_clusres+1):
    model = KMeans(n_clusters=i)
    model.fit(x)
    distances.append(model.inertia_)
plt.plot(range(1, max_clusres+1), distances, marker="o")
```

```python
model = AgglomerativeClustering(n_clusters=3)
model.fit(x)
```

```python
plt.scatter(x["feature1"], x["feature2"], c=model.labels_)
```

```python
distances = pdist(x)
z = linkage(distances, method="ward")
dendrogram(z)
```
